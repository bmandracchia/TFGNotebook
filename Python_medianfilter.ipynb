```python
import numpy as np
from scipy.ndimage import median_filter
import timeit
import statistics
import matplotlib.pyplot as plt
from napari_simpleitk_image_processing import median_filter as median_filter_simpleITK
```

```python
def python_median_filter(image):
    # return median_filter(image, 3, 3)
    return median_filter(image, size=3)

# Run benchmark for different sizes
sizes = [64, 128, 512]
for size in sizes:
    image = np.random.rand(size, size)
    
    # Setup string to create the image in the timeit environment
    setup = f"""
import numpy as np
from scipy.ndimage import median_filter
from __main__ import python_median_filter
image = np.random.rand({size}, {size})
"""
    
    # Run the benchmark
    number = 1  # Number of executions per repeat
    repeat = 100  # Number of repeats
    times = timeit.repeat("python_median_filter(image)", setup=setup, repeat=repeat, number=number)
    
    # Convert to milliseconds
    times_ms = [t * 1000 / number for t in times]
    
    # Calculate statistics
    mean_time = statistics.mean(times_ms)
    std_dev = statistics.stdev(times_ms)
    min_time = min(times_ms)
    max_time = max(times_ms)
    
    print(f"Python - Image size: {size}x{size}")
    print(f"Mean time: {mean_time:.3f} ms")
    print(f"Std Dev:   {std_dev:.3f} ms")
    print(f"Min time:  {min_time:.3f} ms")
    print(f"Max time:  {max_time:.3f} ms")
    print()
    
    # Plot histogram
    plt.figure(figsize=(10, 6))
    plt.hist(times_ms, bins=20, edgecolor='black')
    plt.title(f"Execution Time Distribution for {size}x{size} Image")
    plt.xlabel("Execution Time (ms)")
    plt.ylabel("Frequency")
    plt.show()
```
